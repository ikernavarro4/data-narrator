# datanarrator — Tutorial

**datanarrator** convierte cualquier DataFrame de pandas en un análisis en lenguaje natural.

En este notebook aprenderás a usar todas las funciones de la librería con ejemplos reales.

## 2. Importar la librería

```python
import pandas as pd
from datanarrator import Narrator
```

## 3. Cargar un dataset

Usaremos el dataset del Titanic, uno de los más conocidos en ciencia de datos:

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print(f'Dataset cargado: {df.shape[0]} filas, {df.shape[1]} columnas')
df.head()
```

## 4. Análisis completo con describe()

El método `describe()` genera un análisis completo en lenguaje natural:

```python
n = Narrator(df, lang='es')
print(n.describe())
```

## 5. Resumen ejecutivo con executive_summary()

Para cuando necesitas un resumen rápido de 2-3 oraciones:

```python
print(n.executive_summary())
```

## 6. Solo alertas con alerts_only()

Para ver únicamente los problemas que necesitan atención:

```python
print(n.alerts_only())
```

## 7. Exportar el análisis con export()

Para guardar el análisis en un archivo:

```python
n.export('reporte_titanic.md')

with open('reporte_titanic.md', 'r') as f:
    print(f.read())
```

## 8. Comparar dos datasets con compare()

Para detectar diferencias y data drift entre dos datasets.

Simulamos un dataset de producción con datos diferentes:

```python
df_produccion = df.sample(frac=0.7, random_state=42).copy()
df_produccion['Age'] = df_produccion['Age'] * 1.5

print(n.compare(df_produccion))
```

## 9. Soporte en inglés

Todos los métodos funcionan también en inglés:

```python
n_en = Narrator(df, lang='en')
print(n_en.describe())
```

## 10. Ejemplo con otro dataset — Iris

Probamos con el dataset Iris para mostrar que funciona con cualquier DataFrame:

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df_iris = iris.frame

n_iris = Narrator(df_iris, lang='es')
print(n_iris.describe())
```

| Método | Descripción |
|--------|-------------|
| `describe()` | Análisis completo en lenguaje natural |

## 11. Caso de uso real — Dataset de Diabetes

Usamos el dataset de diabetes de sklearn para mostrar cómo `datanarrator` funciona con datos médicos y cómo `suggest()` detecta problemas de clasificación binaria:

```python
from sklearn.datasets import load_diabetes
import pandas as pd
from datanarrator import Narrator

# Cargamos el dataset de diabetes
diabetes = load_diabetes(as_frame=True)
df_diabetes = diabetes.frame

print('Dataset de Diabetes:')
print(f'{df_diabetes.shape[0]} pacientes, {df_diabetes.shape[1]} variables')
df_diabetes.head()
```

```python
# Análisis completo del dataset médico
n_diabetes = Narrator(df_diabetes, lang='es')
print(n_diabetes.describe())
```

```python
# Sugerencias de modelado para datos médicos
print(n_diabetes.suggest())
```

## 12. Caso de uso real — Precios de Casas en California

Usamos el dataset de precios de casas para mostrar cómo `suggest()` detecta problemas de regresión:

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
from datanarrator import Narrator

# Cargamos el dataset de precios de casas en California
housing = fetch_california_housing(as_frame=True)
df_housing = housing.frame

print('Dataset de Precios de Casas California:')
print(f'{df_housing.shape[0]} casas, {df_housing.shape[1]} variables')
df_housing.head()
```

```python
# Análisis completo del dataset de bienes raíces
n_housing = Narrator(df_housing, lang='es')
print(n_housing.describe())
```

```python
# Sugerencias de modelado — debería detectar regresión
print(n_housing.suggest())
```

```python
# Comparamos el primer 70% de casas vs el 30% restante
# para simular un caso de data drift en producción
df_housing_train = df_housing.iloc[:int(len(df_housing)*0.7)]
df_housing_prod = df_housing.iloc[int(len(df_housing)*0.7):]

n_train = Narrator(df_housing_train, lang='es')
print(n_train.compare(df_housing_prod))
```
